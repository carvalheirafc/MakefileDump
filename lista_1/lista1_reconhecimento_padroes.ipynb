```
import numpy as np
from scipy import stats
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```
X = np.random.random(size=1000)
Y = np.random.random(size=1000)

gauss = np.random.normal(size=1000)
```

```
img, ax = plt.subplots(figsize=(9,7)) 
ax = sns.distplot(gauss)


ax.axvline(gauss.mean(), color='r', label='Mean')
ax.axvline(np.median(gauss), color='b', label='Median')
ax.axvline(gauss.std(), color='g', label='Std')
ax.axvline(gauss.var(), color='lightgreen', label='Var')

ax.legend()
ax.set_title('Variable: Gaussian')
plt.show()
```

```
img, ax = plt.subplots(figsize=(9,7)) 
ax = sns.distplot(X)


ax.axvline(X.mean(), color='r', label='Mean')
ax.axvline(np.median(X), color='b', label='Median')
ax.axvline(X.std(), color='g', label='Std')
ax.axvline(X.var(), color='lightgreen', label='Var')

ax.legend()
ax.set_title('Variable: X')
plt.show()
```

```
img, ax = plt.subplots(figsize=(9,7)) 
ax = sns.distplot(Y)


ax.axvline(Y.mean(), color='r', label='Mean')
ax.axvline(np.median(Y), color='b', label='Median')
ax.axvline(Y.std(), color='g', label='Std')
ax.axvline(Y.var(), color='lightgreen', label='Var')

ax.legend()
ax.set_title('Variable: Y')
plt.show()
```

```
features_df = pd.DataFrame()
features_df['X'] = X
features_df['Y'] = Y
features_df
```

```
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.heatmap(features_df.cov(), annot=True, xticklabels=True)

ax.set_title('Covariance Between X and Y')
plt.show()
```

```
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.heatmap(features_df.corr(), annot=True)

ax.set_title('Correlation Between X and Y')
plt.show()
```
