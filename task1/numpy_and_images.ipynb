```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
original_file = cv2.imread('lena.jpeg')
original_file.shape
# y_range, x_range, 
```

```
def vertical_flip(image):
  #Image rows are inverted
  return image[::-1, :]
  
def horizontal_flip(image):
  #Image column are inverted
  return image[:, ::-1]

def convert_to_rgb(image):
  return image[...,::-1]
```

```
rgb_converted = convert_to_rgb(original_file)
upside_down = vertical_flip(rgb_converted)
plt.imshow(upside_down)
plt.show()

mirror_image = horizontal_flip(rgb_converted)
plt.imshow(mirror_image)
plt.show()
```

```
def blur(a):
    kernel = np.array([[1,2,1], [2,4,2], [1,2,1]])
    kernel = kernel / np.sum(kernel)
    arraylist = []
    for y in range(3):
        temparray = np.copy(a)
        temparray = np.roll(temparray, y - 1, axis=0)
        for x in range(3):
            temparray_X = np.copy(temparray)
            temparray_X = np.roll(temparray_X, x - 1, axis=1)*kernel[y,x]
            arraylist.append(temparray_X)

    arraylist = np.array(arraylist)
    arraylist_sum = np.sum(arraylist, axis=0)
    return arraylist_sum
```

```
blur_img = blur(original_file)
blur_img
```
